```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#### draw a line 
xpoints=np.array([1,2,3])
ypoints=np.array([10,17,14])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,20])
ypoints=np.array([10,200])
plt.plot(xpoints,ypoints,'+')
plt.show()
```

```python
#### we can use +,*,o....
```

```python
xpoints=np.array([3,8,1,10])
ypoints=np.array([5,6,7,9])
plt.plot(xpoints,ypoints,marker='o',ms=100,color='green')
plt.show()
```

```python
### you can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mec='y',mfc='k')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mec='r',mfc='b')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o', ms=20, mec='y', mfc='k')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints, linestyle='dashed', marker='o',ms=20,mec='hotpink',mfc='blue')
plt.show()
```

```python
y1= np.array([3,8,1,10])
y2= np.array([6,2,5,11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```
